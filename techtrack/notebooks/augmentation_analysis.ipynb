```python
# notebooks/augmentation_analysis.ipynb

# Import necessary libraries
import matplotlib.pyplot as plt
import cv2
import os
```

```python

# Import  augmentation functions
from techtrack.rectification.augmentation import (
    horizontal_flip, gaussian_blur, resize_image, rotate_image, adjust_brightness, add_noise
)
```

```python


image_dir = 'logistics'

# Load a sample of images. ensure they're loaded correct
sample_files = os.listdir(image_dir)[:10]  
images = [cv2.imread(os.path.join(image_dir, file)) for file in sample_files]
images = [img for img in images if img is not None]  

if len(images) == 0:
    raise ValueError("No valid images found in the specified directory.")

#  augmentations to apply
augmentations = [
    ('Horizontal Flip', lambda img: horizontal_flip(img, [])[0]),
    ('Gaussian Blur', lambda img: gaussian_blur(img, [])[0]),
    ('Resize', lambda img: resize_image(img, [], (300, 300))[0]),
    ('Rotate 45°', lambda img: rotate_image(img, [], angle=45)[0]),
    ('Adjust Brightness', lambda img: adjust_brightness(img, [], factor=1.5)[0]),
    ('Add Noise', lambda img: add_noise(img, [])[0])
]

# Apply each augmentation
for i, img in enumerate(images):
    plt.figure(figsize=(12, 8))
    plt.subplot(2, 4, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title("Original")
    plt.axis('off')
    
    for j, (name, aug_func) in enumerate(augmentations):
        aug_img = aug_func(img.copy())  #  augmentation
        plt.subplot(2, 4, j + 2)
        plt.imshow(cv2.cvtColor(aug_img, cv2.COLOR_BGR2RGB))
        plt.title(name)
        plt.axis('off')
    
    plt.show()

#  impact of each augmentation
print("Analysis:")
print("- Horizontal Flip: Flipping the image about  horizontal axis ")
print("- Gaussian Blur: Blurring the image using a gaussian function")
print("- Resize: Changing the soze of the image?")
print("- Rotate: Rotating the image by 45 degrees")
print("- Brightness Adjustment: Brighten the Image?")
print("- Noise: Add noise to the image.")

```
